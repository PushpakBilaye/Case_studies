### 10. Total Number of Job Titles Contain Captain

```python
data.JobTitle.sample(50)
```

```python
def findjt(jt):
    if 'captain' in jt.lower():
        return True
    else:
        return False

data['JobTitle'].apply(lambda x:findjt(x)).sum()
```
