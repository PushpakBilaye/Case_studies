# Year wise analysis

### matches played in each year

```python
df['season'].value_counts()
```

### lets analyse ipl season held in 2013. Extract all the details of 2013 season

```python

mean_s=df.groupby(by='season').mean()
max_s=df.groupby(by='season').max()
min_s=df.groupby(by='season').min()
sum_s=df.groupby(by='season').sum()


print("Average details of 2013 \n",mean_s.loc[2013,:])
print("\n Maximum of 2013 \n",max_s.loc[2013,:])
print("\n Minimum of 2013 \n",min_s.loc[2013,:])
print("\n Sum of 2013 \n",sum_s.loc[2013,:])
```

### here we will create yearwise groups (hint: use groupby() function)

```python
s=df.groupby('season')
s.get_group(2013)
```
